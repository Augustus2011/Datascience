##### train_k_fold and test function

```python
#we got list_folf =[f1,f2,f3,f4,f5]
def train_k_fold_and_test(model,list_fold,x_train,y_train):
  acc=[]
  pre=[]
  rec=[]
  roc_auc_list=[]
  f1_macro=[]
  model=model
  for fold in list_fold:
    model.fit(x_train.iloc[fold[0]],y_train.iloc[fold[0]])
    model_pred=model.predict(x_train.iloc[fold[1]])
    acc.append(accuracy_score(y_train.iloc[fold[1]],model_pred))
    pre.append(precision_score(y_train.iloc[fold[1]],model_pred))
    rec.append(recall_score(y_train.iloc[fold[1]],model_pred))
    roc_auc_list.append(roc_auc_score(y_train.iloc[fold[1]],model_pred))
    f1_macro.append(f1_score(y_train.iloc[fold[1]],model_pred,average="binary"))
  return pd.DataFrame({"accuracy":acc,"precision":pre,"recall":rec,"roc_auc":roc_auc_list,"f1_binary":f1_macro}),create_result_df(y_true,test_data_nolabel,model)
```

```python
cat_clf3=catboost.CatBoostClassifier(iterations=30,learning_rate=0.3,depth=6)
result_k_fold,cat3_res_df=train_k_fold_and_test(cat_clf3,list_fold,x_train.drop(["diagnosis"],axis=1),y_train)
%timeit train_k_fold_and_test(cat_clf3,list_fold,x_train.drop(["diagnosis"],axis=1),y_train)
```
